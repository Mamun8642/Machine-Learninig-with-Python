<h1> Feature Transformation </h1>

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('supershops.csv')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True )
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df_norm = df.copy()
model_norm = MinMaxScaler()
```

```python
df_norm['Marketing Spend'] = model_norm.fit_transform(df_norm[['Marketing Spend']])
df_norm['Administration'] = model_norm.fit_transform(df_norm[['Administration']])
df_norm['Transport'] = model_norm.fit_transform(df_norm[['Transport']])
df_norm['Profit'] = model_norm.fit_transform(df_norm[['Profit']])
```

```python
df_norm.dropna()
```

<h2> Standardizatin </h2>

```python
from sklearn.preprocessing import StandardScaler 
```

```python
df_std =  df.copy()
model_std = StandardScaler()
```

```python
df_std['Marketing Spend'] = model_std.fit_transform(df_std[['Marketing Spend']]) # fit() = training the model ; transform() = replace the values
df_std['Administration'] = model_std.fit_transform(df_std[['Administration']])
df_std['Transport'] = model_std.fit_transform(df_std[['Transport']])
df_std['Profit'] = model_std.fit_transform(df_std[['Profit']])
```
